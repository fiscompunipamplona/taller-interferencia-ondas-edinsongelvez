 Representación de las ondas generadas sobre el estanque cuando se deja caer una piedra en un cierto lugar. 

```python
from pylab import plot, imshow, show, xlabel, ylabel
from numpy import loadtxt, array, zeros, sqrt, pi, sin

pix = 500
size = 100. # cm 
step = size / pix # cm/pixel
lda = 5.0 # cm
k = 2.*pi / lda
zeta0 = 1.0 # cm
x1 = 50. # cm
y1 = 50. # cm

x = zeros(pix, float)
y = zeros(pix, float)

for i in range(pix):
    x[i] = step*i
    y[i] = step*i
    
outfile = open("outdata.dat", "w")


for xx in range( len(x) ):
    for yy in range( len(y) ):
        r1 = sqrt( (x1-x[xx])*(x1-x[xx]) + (y1-y[yy])*(y1-y[yy]) ) 
        outfile.write("%0.1f " % (zeta0*sin(r1*k)))
    
    outfile.write("\n")
    
outfile.close()

data = loadtxt("outdata.dat", float)
xlabel("propagacion de la onda en x")
ylabel("propagacion de la onda en y")
imshow(data)
show()
```

# Interferencia lineal de las ondas (vibraciones)

```python
from pylab import plot, imshow, show, xlabel, ylabel
from numpy import loadtxt, array, zeros, sqrt, pi, sin

pix = 500
size = 100. # cm 
step = size / pix # cm/piXel
lda = 5.0 # cm
k = 2.*pi / lda
zeta0 = 1.0 # cm
X0 = float(input("Escriba el punto donde cae la piedra en X:")) # cm
y1 = float(input("NUevamente escriba el punto donde cae la piedra en y:"))

X = zeros(pix, float)
y = zeros(pix, float)

for i in range(pix):
    X[i] = step*i #COnstruyendo el vector X
    y[i] = step*i #COnstruyendo el vector y
    
vibracion = open("outdata.dat", "w")

for yy in range( len(X) ):
    for XX in range( len(y) ):
        R0 = sqrt( (X0-X[XX])*(X0-X[XX]) + (y1-y[yy])*(y1-y[yy]) )
        R1 = sqrt( ((X0+20)-X[XX])*((X0+20)-X[XX]) + (y1-y[yy])*(y1-y[yy]) ) 
        zeta1=zeta0*sin(R0*k)
        zeta2=zeta0*sin(R1*k)
        vibracion.write("%0.1f " % (zeta1))
        vibracion.write("%0.1f " % (zeta2))
    vibracion.write("\n")
    
vibracion.close()

data = loadtxt("outdata.dat", float)
imshow(data,origin="lower",extent=[0,100,0,100])
xlabel("posición de la onda en X cm")
ylabel("posición de la onda en y cm")
imshow(data)
show()

```

# Interferencia lineal de 2 ondas de igual amplitud producida por dos objetos (piedras). 

```python
from pylab import plot, imshow, show, xlabel, ylabel
from numpy import loadtxt, array, zeros, sqrt, pi, sin

pix = 500
size = 100. # cm 
step = size / pix # cm/pixel
lda = 5.0 # cm
k = 2.*pi / lda
zeta0 = 1.0 # cm
X0 = 50. # cm
Y0 = 50. # cm

x = zeros(pix, float)
y = zeros(pix, float)

for i in range(pix):
    x[i] = step*i
    y[i] = step*i
    
outfile = open("outdata.dat", "w")

for xx in range( len(y) ):
    for yy in range( len(x) ):
        R0 = sqrt( (X0-x[xx])*(X0-x[xx]) + (Y0-y[yy])*(Y0-y[yy]) )
        R1 = sqrt( ((X0+20)-x[xx])*((X0+20)-x[xx]) + (Y0-y[yy])*(Y0-y[yy]) ) 
        zeta1=zeta0*sin(R0*k)
        zeta2=zeta0*sin(R1*k)
        outfile.write("%0.1f " % (zeta1+zeta2))
    
    outfile.write("\n")
    
outfile.close()

data = loadtxt("outdata.dat", float)
xlabel("posición de la onda en x cm ")
ylabel("posición de la onda en y cm ")
imshow(data)
show()
```
